```python
from qiskit import *
import matplotlib 
import numpy as np
```

```python
qc=QuantumCircuit(2,name='oracle')
qc.cz(0,1)
qc.cx(0,1)
qc.to_gate()
qc.draw(output="mpl")
```

```python
backend=Aer.get_backend('statevector_simulator')
grover_circuit=QuantumCircuit(2,2)# (quantumregister,classicalregister)

grover_circuit.h([0,1])# superpostion of all possible state

grover_circuit.append(qc,[0,1])
grover_circuit.draw(output="mpl")
```

```python
job=execute(grover_circuit,backend)
result=job.result()
```

```python
statevector=result.get_statevector()
np.around(statevector,2)
```

```python
ref=QuantumCircuit(2,name='reflection')
ref.h([0,1])
ref.z([0,1])
ref.cz(0,1)
ref.h([0,1])
ref.to_gate()
ref.draw(output='mpl')
```

```python
backend=Aer.get_backend('qasm_simulator')
#circuit for measuring |11> state
grover_circ=QuantumCircuit(2,2)
grover_circ.h([0,1])
grover_circ.append(qc,[0,1])
grover_circ.append(ref,[0,1])
grover_circ.measure([0,1],[0,1])
grover_circ.draw(output='mpl')
```

```python
job=execute(grover_circ,backend,shots=1000)
job.result().get_counts()
```

```python
#for three qubit system with calculation of N

#grover for |101>
def oracle(n):
    qc=QuantumCircuit(n,name="oracle")
    qc.x(1)
    qc.ccx(0,1,3)
    qc.ccx(2,3,4)
    qc.ccx(0,1,3)
    qc.x(1)
    qc.to_gate()
    qc.draw('mpl')
    
    return qc
```

```python
def ref(n):
    qc=QuantumCircuit(n,name="Ref")
    
    for qb in range(n-2):
        qc.h(qb)
        
    for i in range(n-2):
        qc.x(i)
        
    qc.ccx(0,1,3)
    qc.ccx(2,3,4)
    qc.ccx(0,1,3)
    
    for i in range(n-2):
        qc.x(i)
        
    for qb in range(n-2):
        qc.h(qb)
    
    return qc
```

```python
n=5
gr=QuantumCircuit(n,n-2)

r=int(np.floor(np.pi/4*np.sqrt(2**(n-2))))

gr.h(range(n-2))

gr.x(n-1)
gr.h(n-1)

for j in range(r):
    gr.append(oracle(n),range(n))
    gr.append(ref(n),range(n))
    
gr.measure(range(n-2),range(n-2))

gr.draw('mpl')
```

```python
from qiskit.visualization import plot_histogram
backend=Aer.get_backend("qasm_simulator")
job=execute(gr,backend,shota=100)
counts=job.result().get_counts()

plot_histogram(counts)
```
